<h1>Ploting</h1>

```python
time_stamp = pd.Series(np.random.randn(500), index = pd.date_range("1/1/2022", periods = 500))
time_stamp.cumsum()
time_stamp.plot()
```

```python
# dataframe plotting
dataframe = pd.DataFrame(np.random.randn(500, 4), index = time_stamp.index, columns = ['A', 'B', 'C', 'D'])
dataframe = dataframe.cumsum()

plt.figure() 
dataframe.plot()
plt.legend(loc = "best")
```

<h1>Gerando outputs</h1>

```python
# gerando o arquivo excel a partir do dataframe
dataframe.to_excel("test-dataframe-output.xlsx", sheet_name="Sheet1")
```

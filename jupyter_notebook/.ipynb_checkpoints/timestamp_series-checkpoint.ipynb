<h1>Trabalhando com Series temporais</h1>

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
time_range = pd.date_range('1/1/2022', periods = 50, freq = 'S')
```

```python
time_stamp = pd.Series(np.random.randint(0, 500, len(time_range)), index = time_range)
```

```python
time_stamp
```

<h1>Ploting</h1>

```python
time_stamp.cumsum()
time_stamp.plot()
```

```python
# dataframe plotting
dataframe = pd.DataFrame(np.random.randn(500, 4), index = time_stamp.index, columns = ['A', 'B', 'C', 'D'])
dataframe = dataframe.cumsum()

plt.figure() 
dataframe.plot()
plt.legend(loc = "best")
```

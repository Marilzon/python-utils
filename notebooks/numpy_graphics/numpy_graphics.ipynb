<h1>Graficos com o NumPy</h1>

```python
import matplotlib.pyplot as plt
import numpy as np

%matlpotlib inline
```

```python
plt.scatter(np.arange(50), np.random.randn(50))
plt.show()
```

```python
figure = plt.figure()
axis1 = figure.add_subplot(1, 2, 1)
axis1.plot(np.random.randn(50), color="red")

axis2 = figure.add_subplot(1, 2, 2)
axis2.scatter(np.arange(50), np.random.randn(50))
plt.show()
```

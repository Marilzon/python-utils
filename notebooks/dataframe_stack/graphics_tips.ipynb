<h1>Grafico de barras</h1>

```python
x = [2, 4, 6, 8, 10]
y = [6, 7, 8, 2, 4]
```

```python
plt.bar(x, y, label = "Barras", color = "green")
plt.legend()
plt.show()
```

```python
ages = np.random.randint(120, size=20)
ages
```

```python
list_compreention_ages = [item for item in range(len(ages))]
```

```python
# grafico de quantidades de pessoas por idade
plt.bar(list_compreention_ages, ages)
plt.show()
```

<h1>Grafico Histograma</h1>

```python
bins = np.arange(start = 0, stop = 131, step = 10)
bins
```

```python
plt.hist(ages, bins, histtype = "bar", rwidth = 0.8)
plt.show()
```

```python
plt.hist(ages, bins, histtype = "stepfilled", rwidth = 0.8)
plt.show()
```

<h1>Grafico Stack Plots</h1>

```python
days = [1, 2, 10, 13, 28]
sleep = [3, 4, 5, 7, 11]
eat = [2, 3, 5, 8, 15]
work = [1, 15, 20, 30, 31]
workout = [1, 10, 20, 30, 31]
```

```python
plt.stackplot(
    days, 
    sleep, 
    eat, 
    work, 
    workout, 
    colors = ["m", "c", "r", "g", "b"]
)
plt.show()
```

<h1>Pizza Chart(Pie Chart)</h1>

```python
pies = [7, 2, 2, 13]
steps = ["eat", "sleep", "work", "workout"]
columns = ["c", "m", "r", "k"]
```

```python
plt.pie(pies, labels = steps, colors = columns, startangle = 90, shadow = True, explode = (0, 0.1, 0, 0))
plt.show()
```

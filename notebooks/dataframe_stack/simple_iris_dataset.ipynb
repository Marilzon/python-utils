<h1>Trabalhando com datasets no numpy</h1>

```python
import os
import numpy as np
from matplotlib import pyplot as plt

filename = os.path.join('iris.csv')
```

```python
!more iris.csv
```

```python
# carregando o dataset para dentro de um array
file = np.loadtxt(
    filename, 
    delimiter = ',', 
    usecols = (0,1,2,3), 
    skiprows = 1
)
print(file)
```

```python
type(file)
```

```python
# gerando um plot a partir de um arquivo com numpy
alpha, beta = np.loadtxt(
    filename, 
    delimiter = ',', 
    usecols = (0,1), 
    skiprows = 1, 
    unpack = True
)
plt.show(
    plt.plot(
        alpha, 
        beta,
        'o',
        markersize = 8,
        alpha = 0.75
    )
)
```
